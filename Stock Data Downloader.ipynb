```python
import yfinance as yf
import pandas as pd
import numpy as np
from scipy.stats import skew, kurtosis
import matplotlib.pyplot as plt

# --- 1) Define universe and download data ---
# Define the list of cryptocurrency tickers to include in the analysis
tickers = [
    "BTC-USD","ETH-USD","XRP-USD","USDT-USD","BNB-USD",
    "SOL-USD","USDC-USD","DOGE-USD","TRX-USD","ADA-USD",
    "HYPE-USD","XLM-USD","SUI-USD","LINK-USD","HBAR-USD",
    "BCH-USD","AVAX-USD","SHIB-USD","LTC-USD","LEO-USD",
    "TON-USD","DOT-USD","UNI-USD","XMR-USD","BGB-USD"
]

# Define the date range for data download
# Download data from the start of 2014 up to today
# Set group_by to False to avoid a future warning and handle data structure manually
endDate = pd.to_datetime('today').strftime('%Y-%m-%d')
data = yf.download(tickers, start="2010-01-01", end=endDate, group_by=False)

# --- 2) Build signals per coin ---
signals = {}
for t in tickers:
    # Select data for the current ticker and drop rows with any missing values
    # Access data using the ticker as the first level of columns
    df = data[t].copy().dropna()

    # Calculate technical indicators and filters
    # Breakout signal: 20-period rolling maximum of the 'High', shifted by one day
    df['HH250'] = df['High'].rolling(20).max().shift(1)

    # Exponential Moving Averages (EMAs)
    # 5-period EMA of the 'Close' price
    df['EMA5']  = df['Close'].ewm(span=5, adjust=False).mean()
    # Previous day's 5-period EMA
    df['EMA5_p'] = df['EMA5'].shift(1)

    # Bitcoin risk-on filter: True if Bitcoin's close is above its 20-period EMA
    # Get Bitcoin's closing price from the main data DataFrame
    btc = data['BTC-USD']['Close']
    # Calculate Bitcoin's 20-period EMA
    btc_ema20 = btc.ewm(span=20, adjust=False).mean()
    # Align Bitcoin's RiskOn status with the current coin's index
    df['RiskOn'] = df.index.isin(btc.index) & (btc.loc[df.index].gt(btc_ema20.loc[df.index]))


    # Generate trading signal: 1=entry, -1=exit, 0=flat
    df['Signal'] = 0
    # Entry signal conditions:
    # 1. Current Close is above the previous day's 20-period Highest High (Breakout)
    # 2. Current Close is above 0.20 (Liquidity filter)
    # 3. Current Volume is above 10,000 (Liquidity filter)
    # 4. Current 5-period EMA is above the previous day's 5-period EMA (Upward EMA trend)
    # 5. Bitcoin RiskOn filter is True
    df.loc[
        (df['Close'] > df['HH250']) &
        (df['Close'] > 0.20) &
        (df['Volume'] > 10_000) &
        (df['EMA5'] > df['EMA5_p']) &
        df['RiskOn'],
        'Signal'
    ] = 1

    # Exit signal condition:
    # Current 5-period EMA is below the previous day's 5-period EMA (Downward EMA trend)
    df.loc[df['EMA5'] < df['EMA5_p'], 'Signal'] = -1

    # Store the DataFrame with signals for the current ticker
    signals[t] = df

# --- 3) Extract trades across all coins ---
trade_details = []
for t, df in signals.items():
    df = df.copy()
    # Get the previous day's signal for identifying signal changes
    df['Sig_p'] = df['Signal'].shift(1).fillna(0)
    # Identify entry points (Signal changes to 1)
    entries = df[(df['Signal']==1) & (df['Sig_p']!=1)]
    # Identify exit points (Signal becomes -1) - consider all -1 signals as potential exits
    exits   = df[(df['Signal']==-1)]

    i = 0 # index for entries
    j = 0 # index for exits

    # Iterate through entries to find corresponding exits or identify open trades
    while i < len(entries):
        # Get the current entry's index and row data
        e_idx, e_row = entries.iloc[i].name, entries.iloc[i]
        # Find the first exit signal that occurs after the current entry date
        next_exit = exits[exits.index > e_idx]

        # Check if a subsequent exit was found
        if not next_exit.empty:
            # Get the first subsequent exit's index and row data
            x_idx, x_row = next_exit.iloc[0].name, next_exit.iloc[0]
            # Calculate the return for the closed trade
            ret = (x_row['Close'] - e_row['Close']) / e_row['Close']
            # Append the details of the closed trade
            trade_details.append({
                'Ticker': t,
                'Buy Date': e_idx.strftime('%Y-%m-%d'),
                'Buy Price': e_row['Close'],
                'Sell Date': x_idx.strftime('%Y-%m-%d'),
                'Sell Price': x_row['Close'],
                'Return': ret,
                'Status': 'Closed' # Mark as a closed trade
            })
            # Advance the exit index past the one used (not strictly necessary in this while loop structure, but kept for clarity)
            j = exits.index.get_loc(x_idx) + 1
        else:
            # Handle open trades: entry with no subsequent exit found
            # Use the last available date and price as the "sell" details for open positions
            last_date = df.index[-1]
            last_price = df['Close'].iloc[-1]
            # Calculate the current unrealized return for the open trade
            ret = (last_price - e_row['Close']) / e_row['Close']
            # Append the details of the open trade
            trade_details.append({
                'Ticker': t,
                'Buy Date': e_idx.strftime('%Y-%m-%d'),
                'Buy Price': e_row['Close'],
                'Sell Date': last_date.strftime('%Y-%m-%d'), # Use last date as sell date
                'Sell Price': last_price, # Use last price as sell price
                'Return': ret,
                'Status': 'Open' # Mark as an open trade
            })
        # Move to the next entry
        i += 1

# Convert the list of trade details into a pandas DataFrame
trades_df = pd.DataFrame(trade_details)

# --- 4) Tradelevel statistics ---
# Calculate statistics based only on closed trades for traditional metrics
closed_trades_df = trades_df[trades_df['Status'] == 'Closed'].copy()
trades = closed_trades_df['Return']
wins   = trades[trades > 0]
losses = trades[trades < 0]

# Calculate various trade performance metrics
win_pct       = len(wins) / len(trades) * 100 if len(trades) > 0 else 0
avg_profit    = wins.mean() * 100 if len(wins) > 0 else 0
avg_loss      = losses.mean() * 100 if len(losses) > 0 else 0
expectancy    = trades.mean() * 100 if len(trades) > 0 else 0
# Profit factor is total winnings divided by total losses (absolute value)
profit_factor = wins.sum() / abs(losses.sum()) if losses.sum() != 0 else np.inf if wins.sum() > 0 else 0

# --- 5) Portfolio daily returns (equalweight among in signal coins) ---
# Create a DataFrame of closing prices for all tickers
# Access Close prices using the ticker as the first level of columns
price_df = pd.DataFrame({t: data[t]['Close'] for t in tickers})
# Calculate next-day percentage returns, explicitly setting fill_method to None
ret_df   = price_df.pct_change(fill_method=None).shift(-1)   # nextday returns
# Create a DataFrame of daily positions based on signals (1 if in trade, 0 if out)
pos_df   = pd.DataFrame({t: signals[t]['Signal'] for t in tickers})
# Shift positions by one day to align with next-day returns and ensure no lookahead bias
pos_df   = pos_df.shift(1).fillna(0).clip(lower=0) # Clip to ensure positions are 0 or 1
# Calculate daily portfolio return (sum of returns of held assets divided by number of held assets)
port_ret = (ret_df * pos_df).sum(axis=1) / pos_df.sum(axis=1)
# Fill any days with no positions with a 0 return
port_ret = port_ret.fillna(0)
# Calculate the cumulative equity curve
equity   = (1 + port_ret).cumprod()

# --- 6) Portfoliolevel statistics ---
trading_days = 252 # Approximate number of trading days in a year
# Calculate the number of years covered by the data
years        = len(port_ret) / trading_days
# Calculate Compound Annual Growth Rate (CAGR)
cagr         = equity.iloc[-1] ** (1/years) - 1 if len(equity) > 1 else 0
# Calculate Sharpe Ratio (using risk-free rate of 0)
sharpe       = port_ret.mean() / port_ret.std() * np.sqrt(trading_days) if port_ret.std() != 0 else np.nan
# Calculate Maximum Drawdown
roll_max     = equity.cummax()
max_dd       = ((equity - roll_max) / roll_max).min() if len(equity) > 0 else 0

# --- 7) Summary and Output ---
# Create a Series to hold the summary statistics
stats = pd.Series({
    "Number of Trades (Closed)": len(closed_trades_df),
    "Number of Trades (Open)": len(trades_df[trades_df['Status'] == 'Open']),
    "Total Number of Trades": len(trades_df),
    "Win % (Closed Trades)": f"{win_pct:.2f}%",
    "Avg Profit % (Closed Trades)": f"{avg_profit:.2f}%",
    "Avg Loss % (Closed Trades)": f"{avg_loss:.2f}%",
    "Expectancy % (Closed Trades)": f"{expectancy:.2f}%",
    "Profit Factor (Closed Trades)": f"{profit_factor:.2f}",
    "Annual Return (CAGR)": f"{cagr*100:.2f}%",
    "Sharpe Ratio": f"{sharpe:.2f}" if not pd.isna(sharpe) else "N/A",
    "Max Drawdown %": f"{max_dd*100:.2f}%"
})
print("--- Summary Statistics ---")
print(stats)

# Print details of open positions
print("\n--- Open Positions ---")
open_trades = trades_df[trades_df['Status'] == 'Open']
if not open_trades.empty:
    display(open_trades[['Ticker', 'Buy Date', 'Buy Price', 'Sell Date', 'Sell Price', 'Return']])
else:
    print("No open positions.")

# Print latest signals from 2024 and 2025
print("\n--- Latest Signals (2024 and 2025) ---")
latest_signals = {}
for t, df in signals.items():
    # Filter for signals in the years 2024 and 2025
    recent_signals = df.loc['2024':'2025']
    if not recent_signals.empty:
        # Get the last signal row for the current ticker
        latest_signal_row = recent_signals.iloc[-1]
        # Store the latest signal details
        latest_signals[t] = {
            'Date': latest_signal_row.name.strftime('%Y-%m-%d'),
            'Signal': latest_signal_row['Signal'],
            'Close': latest_signal_row['Close']
        }

# Display the latest signals in a DataFrame if any were found
if latest_signals:
    latest_signals_df = pd.DataFrame.from_dict(latest_signals, orient='index')
    latest_signals_df.index.name = 'Ticker'
    display(latest_signals_df)
else:
    print("No signals found in 2024 and 2025.")

# --- Data Visualization ---
# Plot the equity curve
plt.figure(figsize=(12, 6))
plt.plot(equity)
plt.xlabel('Date')
plt.ylabel('Equity')
plt.title('Trading Strategy Equity Curve')
plt.grid(True)
plt.show()

# Plot histogram for returns of closed trades
plt.figure(figsize=(12, 6))
plt.hist(closed_trades_df['Return'], bins=50, edgecolor='black')
plt.xlabel('Return')
plt.ylabel('Frequency')
plt.title('Distribution of Closed Trade Returns')
plt.grid(True)
plt.show()

# --- 3) Extract trades across all coins ---
trade_details = []
open_positions   = {}       # key: ticker, value: dict with entry info
max_open_trades  = 10       #   <-- HARD CAP
max_wgt_per_trade = 0.10    #   <-- 10 % of equity

equity_curve = [1.]         # running equity to size trades
equity_dates = [signals[tickers[0]].index[0]]

for t, df in signals.items():
    df = df.copy()
    df['Sig_p'] = df['Signal'].shift(1).fillna(0)
    entries = df[(df['Signal']==1) & (df['Sig_p']!=1)]
    exits   = df[(df['Signal']==-1)]

    for e_idx, e_row in entries.iterrows():

        # ---------- 1) handle exits that happened BEFORE this entry ----------
        #   (close positions first to free slots)
        exited_now = []
        for pos_tkr, pos in list(open_positions.items()):
            # check if an exit signal for that positions coin occurred
            coin_exits = signals[pos_tkr][signals[pos_tkr]['Signal']==-1]
            later_exit = coin_exits[coin_exits.index > pos['Buy Date']]
            if not later_exit.empty and later_exit.index[0] <= e_idx:
                x_idx = later_exit.index[0]
                x_price = later_exit.iloc[0]['Close']
                ret = (x_price - pos['Buy Price'])/pos['Buy Price']
                trade_details.append({
                    'Ticker': pos_tkr, 'Buy Date': pos['Buy Date'].strftime('%Y-%m-%d'),
                    'Buy Price': pos['Buy Price'],
                    'Sell Date': x_idx.strftime('%Y-%m-%d'),
                    'Sell Price': x_price, 'Return': ret, 'Status': 'Closed'})
                # update equity
                equity_curve.append(equity_curve[-1]*(1+ret))
                equity_dates.append(x_idx)
                exited_now.append(pos_tkr)

        # remove closed positions from the dict
        for pos_tkr in exited_now:
            open_positions.pop(pos_tkr, None)

        # ---------- 2) apply caps BEFORE opening a new trade ----------
        # a) max 10 concurrent positions
        if len(open_positions) >= max_open_trades:
            continue

        # b) max 10 % of equity
        current_equity = equity_curve[-1]
        notional_size  = e_row['Close']              # 1 coin position
        if notional_size/current_equity > max_wgt_per_trade:
            # scale position size down to 10 % of equity
            position_units = (max_wgt_per_trade*current_equity)/e_row['Close']
        else:
            position_units = 1

        # open the position
        open_positions[t] = {'Buy Date': e_idx, 'Buy Price': e_row['Close'],
                             'Units': position_units}

    # ---------- 3) handle exits after the final entry loop ----------
    # (do this once per coin after processing all entries)
    for pos_tkr, pos in list(open_positions.items()):
        # check for exit signals beyond last entry date up to df end
        coin_exits = signals[pos_tkr][signals[pos_tkr]['Signal']==-1]
        later_exit = coin_exits[coin_exits.index > pos['Buy Date']]
        if not later_exit.empty:
            x_idx = later_exit.index[0]
            x_price = later_exit.iloc[0]['Close']
            ret = (x_price - pos['Buy Price'])/pos['Buy Price']
            trade_details.append({
                'Ticker': pos_tkr, 'Buy Date': pos['Buy Date'].strftime('%Y-%m-%d'),
                'Buy Price': pos['Buy Price'],
                'Sell Date': x_idx.strftime('%Y-%m-%d'),
                'Sell Price': x_price, 'Return': ret, 'Status': 'Closed'})
            equity_curve.append(equity_curve[-1]*(1+ret))
            equity_dates.append(x_idx)
            open_positions.pop(pos_tkr, None)

# after iterating all coins, mark any leftovers as open
for pos_tkr, pos in open_positions.items():
    last_date  = signals[pos_tkr].index[-1]
    last_price = signals[pos_tkr]['Close'].iloc[-1]
    ret = (last_price-pos['Buy Price'])/pos['Buy Price']
    trade_details.append({
        'Ticker': pos_tkr, 'Buy Date': pos['Buy Date'].strftime('%Y-%m-%d'),
        'Buy Price': pos['Buy Price'],
        'Sell Date': last_date.strftime('%Y-%m-%d'),
        'Sell Price': last_price, 'Return': ret, 'Status': 'Open'})
```

```python
import yfinance as yf
import pandas as pd
import numpy as np
from scipy import stats # For Sortino Ratio
import matplotlib.pyplot as plt

# --- 1) Define universe and download data ---
tickers = [
    "BTC-USD", "ETH-USD", "XRP-USD", "USDT-USD", "BNB-USD",
    "SOL-USD", "USDC-USD", "DOGE-USD", "TRX-USD", "ADA-USD",
    "HYPE-USD", "XLM-USD", "SUI-USD", "LINK-USD", "HBAR-USD",
    "BCH-USD", "AVAX-USD", "SHIB-USD", "LTC-USD", "LEO-USD",
    "TON-USD", "DOT-USD", "UNI-USD", "XMR-USD", "BGB-USD"
]
endDate = pd.to_datetime('today').strftime('%Y-%m-%d')
data = yf.download(tickers, start="2010-01-01", end=endDate, group_by=False)

# --- 2) Build signals per coin ---
signals = {}
for t in tickers:
    # Select data for the current ticker and drop rows with any missing values
    # Access data using the ticker as the first level of columns
    df = data[t].copy()

    # Ensure required columns exist and drop rows with NaNs in them
    required_cols = ['Open', 'High', 'Low', 'Close', 'Volume']
    if not all(col in df.columns for col in required_cols):
        print(f"Skipping signal calculation for {t}: Missing one or more required columns ({required_cols}).")
        signals[t] = pd.DataFrame() # Store empty DataFrame
        continue
    df.dropna(subset=required_cols, inplace=True)

    if df.empty:
        print(f"Skipping signal calculation for {t}: No valid data after dropping NaNs.")
        signals[t] = pd.DataFrame() # Store empty DataFrame
        continue

    # Ensure there are enough data points to calculate rolling metrics
    if len(df) < 20: # HH250 requires 20 periods
         print(f"Skipping signal calculation for {t}: Not enough data points ({len(df)}).")
         signals[t] = pd.DataFrame() # Store empty DataFrame
         continue


    # Calculate technical indicators and filters
    # Breakout signal: 20-period rolling maximum of the 'High', shifted by one day
    # Ensure 'High' column exists before calculating rolling max
    if 'High' not in df.columns:
        print(f"Skipping signal calculation for {t}: 'High' column missing after dropna.")
        signals[t] = pd.DataFrame() # Store empty DataFrame
        continue
    df['HH250'] = df['High'].rolling(20).max().shift(1)

    # Exponential Moving Averages (EMAs)
    # Ensure 'Close' column exists before calculating EMAs
    if 'Close' not in df.columns:
         print(f"Skipping signal calculation for {t}: 'Close' column missing after dropna.")
         signals[t] = pd.DataFrame() # Store empty DataFrame
         continue
    df['EMA5'] = df['Close'].ewm(span=5, adjust=False).mean() # 5-period EMA
    df['EMA5_p'] = df['EMA5'].shift(1)

    # Ensure BTC-USD data is available for RiskOn filter
    # Also check if BTC-USD data has enough points and the 'Close' column
    btc_risk_on = True # Default to True if BTC data is unavailable or insufficient
    if 'BTC-USD' in data.columns:
        btc_data = data['BTC-USD'].copy().dropna(subset=['Close'])
        if len(btc_data) > 20 and 'Close' in btc_data.columns:
            btc_close_series = btc_data['Close']
            btc_ema20_series = btc_close_series.ewm(span=20, adjust=False).mean()
            # Align Bitcoin's RiskOn status with the current coin's index
            aligned_btc_close = btc_close_series.reindex(df.index, method='ffill')
            aligned_btc_ema20 = btc_ema20_series.reindex(df.index, method='ffill')
            df['RiskOn'] = (aligned_btc_close > aligned_btc_ema20).fillna(False)
            btc_risk_on = False # RiskOn filter is active
        else:
             print(f"Warning: BTC-USD data insufficient or missing 'Close'. RiskOn filter disabled for {t}.")
             df['RiskOn'] = True # Default to True if Bitcoin data is unavailable or insufficient

    else:
        print(f"Warning: BTC-USD data column not found. RiskOn filter disabled for {t}.")
        df['RiskOn'] = True # Default to True if Bitcoin data is unavailable


    # Generate trading signal: 1=entry, -1=exit, 0=flat
    df['Signal'] = 0

    # Define entry conditions, ensuring all necessary columns exist and are not NaN
    entry_conditions = (df['Close'] > df.get('HH250', pd.Series(np.nan, index=df.index))) & \
                       (df['Close'] > 0.20) & \
                       (df['Volume'] > 10_000) & \
                       (df.get('EMA5', pd.Series(np.nan, index=df.index)) > df.get('EMA5_p', pd.Series(np.nan, index=df.index))) & \
                       df.get('RiskOn', pd.Series(True, index=df.index)) # Default RiskOn to True if column is missing

    # Apply entry signal where conditions are True and relevant columns are not NaN
    df.loc[entry_conditions.fillna(False), 'Signal'] = 1

    # Exit signal condition:
    # Current 5-period EMA is below the previous day's 5-period EMA (Downward EMA trend)
    # Ensure EMA5 and EMA5_p columns exist and are not NaN for exit condition
    exit_conditions = (df.get('EMA5', pd.Series(np.nan, index=df.index)) < df.get('EMA5_p', pd.Series(np.nan, index=df.index)))
    # Apply exit signal where conditions are True and relevant columns are not NaN
    df.loc[exit_conditions.fillna(False), 'Signal'] = -1


    # Store the DataFrame with signals for the current ticker
    signals[t] = df

# --- 3) Portfolio daily returns with Constraints and Margin ---
# Create a DataFrame of closing prices for all tickers
price_df = pd.DataFrame({t: data[t]['Close'] for t in tickers if t in data.columns})
# Calculate next-day percentage returns
ret_df = price_df.pct_change(fill_method=None).shift(-1)

# Create a DataFrame of daily signals
signal_df = pd.DataFrame({t: signals[t]['Signal'] for t in tickers if t in signals and not signals[t].empty and 'Signal' in signals[t].columns})


# --- Implement Portfolio with Constraints and Margin ---
max_open_trades = 15
max_weight_per_trade = 0.10
margin = 1.50  # 150% margin (1.50 = 150%)

# Initialize portfolio tracking structures
portfolio_value = [1.0]  # Start with portfolio value 1
portfolio_dates = [price_df.index[0]]  # Start date
portfolio_weights = pd.DataFrame(0.0, index=price_df.index, columns=tickers)  # Daily weights

# List to store detailed trade information
all_trade_details = []

# Simulate the portfolio day by day
current_positions = {}  # key: ticker, value: {'entry_date': ..., 'entry_price': ..., 'weight': ...}

# Iterate through dates (from the second date to have returns)
for date in price_df.index[1:]:  # Start from the second date
    prev_date = price_df.index[price_df.index.get_loc(date) - 1]

    # 1. Update portfolio value based on yesterday's positions and today's returns
    # Get yesterday's weights (used to calculate yesterday's position)
    # Ensure prev_date exists in portfolio_weights index
    if prev_date in portfolio_weights.index:
        prev_weights = portfolio_weights.loc[prev_date]
        # Get today's returns
        today_returns = ret_df.loc[date]

        # Calculate PnL for yesterday's positions
        pnl = (prev_weights * today_returns).sum()
        new_value = portfolio_value[-1] * (1 + pnl)
        portfolio_value.append(new_value)
        portfolio_dates.append(date)
    else:
         # If prev_date is not in portfolio_weights, just carry over the last value and date
         portfolio_value.append(portfolio_value[-1])
         portfolio_dates.append(date)
         prev_weights = pd.Series(0.0, index=tickers) # Assume no positions if no previous weights


    # 2. Handle Exits for current positions based on today's signal
    # Check if any held coin signals an exit today
    exited_tickers = []
    for ticker in list(current_positions.keys()):
        # Ensure the date exists in the signal data for the ticker and the signal column exists
        if date in signal_df.index and ticker in signal_df.columns and signal_df.loc[date, ticker] == -1:
            # Exit signal received, record trade and remove position
            entry_info = current_positions[ticker]
            # Ensure the date exists in price_df for the ticker and the Close price is available
            if date in price_df.index and ticker in price_df.columns and pd.notna(price_df.loc[date, ticker]):
                exit_price = price_df.loc[date, ticker]
                trade_return = (exit_price - entry_info['entry_price']) / entry_info['entry_price']

                all_trade_details.append({
                    'Ticker': ticker,
                    'Buy Date': entry_info['entry_date'].strftime('%Y-%m-%d'),
                    'Buy Price': entry_info['entry_price'],
                    'Sell Date': date.strftime('%Y-%m-%d'),
                    'Sell Price': exit_price,
                    'Return': trade_return,
                    'Status': 'Closed'
                })

                exited_tickers.append(ticker)
            else:
                 print(f"Warning: Price data missing for {ticker} on {date}. Cannot process exit trade.")


    # Remove exited positions
    for ticker in exited_tickers:
        current_positions.pop(ticker, None)


    # 3. Handle Entries based on today's signal, respecting constraints
    # Get tickers signaling entry today (Signal = 1)
    if date in signal_df.index:
        candidates = signal_df.loc[date][signal_df.loc[date] == 1].index.tolist()
    else:
        candidates = []

    # Try to enter new positions
    for ticker in candidates:
        # Skip if already holding
        if ticker in current_positions:
            continue

        # Constraint 1: Max Open Trades
        if len(current_positions) >= max_open_trades:
            continue  # Cannot open more positions

        # Constraint 2: Max Weight per Trade (relative to total portfolio value + margin)
        # Use the latest calculated portfolio value *before* today's trades
        current_portfolio_value = portfolio_value[-2] if len(portfolio_value) > 1 else 1.0
        total_available_capital = current_portfolio_value * (1.0 + margin)

        # Calculate the intended notional value for this trade
        intended_notional_value = max_weight_per_trade * total_available_capital

        # Calculate the current total notional value of open positions
        current_total_notional = sum(
            current_positions[pos_tkr]['weight'] * total_available_capital
            for pos_tkr in current_positions.keys()
        )
        # Calculate the remaining notional value available for new allocations
        available_notional_for_allocation = total_available_capital - current_total_notional

        # Ensure the entry price is available and valid
        if date in price_df.index and ticker in price_df.columns and not pd.isna(price_df.loc[date, ticker]) and price_df.loc[date, ticker] > 0:
             entry_price = price_df.loc[date, ticker]

             # Check if we can allocate the intended notional value
             # Also ensure the entry price allows for a position > 0
             if intended_notional_value <= available_notional_for_allocation and intended_notional_value / entry_price > 0:

                 # If we pass constraints, enter the position
                 # The weight is the intended notional value as a fraction of total available capital
                 actual_weight = intended_notional_value / total_available_capital

                 current_positions[ticker] = {
                     'entry_date': date, # Store the entry date
                     'entry_price': entry_price,
                     'weight': actual_weight # Store the weight relative to total available capital
                 }

                 # Record the new trade as 'Open' immediately
                 all_trade_details.append({
                     'Ticker': ticker,
                     'Buy Date': date.strftime('%Y-%m-%d'),
                     'Buy Price': entry_price,
                     'Sell Date': None, # Will be updated when closed
                     'Sell Price': None, # Will be updated when closed
                     'Return': None, # Will be updated when closed or at the end
                     'Status': 'Open'
                 })
             else:
                 # print(f"Cannot enter {ticker} on {date}: Not enough available allocation value.")
                 pass # Cannot enter with required allocation value
        else:
             # print(f"Cannot enter {ticker} on {date}: Entry price not available or invalid.")
             pass # Entry price not available or valid


    # 4. Record today's weights for use tomorrow
    # Weights are stored relative to the total available capital (portfolio value + margin)
    # This needs to be based on the *current* open positions after today's entries/exits
    for ticker in tickers:
        # Ensure the date exists in portfolio_weights index before assigning
        if date in portfolio_weights.index:
            portfolio_weights.loc[date, ticker] = current_positions.get(ticker, {}).get('weight', 0.0)
        else:
             # Handle case where date is not in index (shouldn't happen with price_df index iteration)
             pass


# --- Handle Open Positions at the end of the backtest ---
# These were already added when initiated. We need to update their 'current' return for display.
# Ensure price_df is not empty before getting the last date
if not price_df.empty:
    last_date = price_df.index[-1]
    for trade in all_trade_details:
        if trade['Status'] == 'Open' and trade['Sell Date'] is None:
            ticker = trade['Ticker']
            buy_date = pd.to_datetime(trade['Buy Date'])
            # Find the latest available price for this ticker after the buy date
            if ticker in price_df.columns:
                recent_prices = price_df.loc[price_df.index >= buy_date, ticker]
                if not recent_prices.empty and pd.notna(recent_prices.iloc[-1]):
                    last_price = recent_prices.iloc[-1]
                    trade_return = (last_price - trade['Buy Price']) / trade['Buy Price']
                    # Update the trade details for display
                    trade['Sell Date'] = recent_prices.index[-1].strftime('%Y-%m-%d') # Use the date of the last price
                    trade['Sell Price'] = last_price
                    trade['Return'] = trade_return
                else:
                    # Handle cases where no price data exists after the buy date
                    trade['Sell Date'] = last_date.strftime('%Y-%m-%d') # Use the last date of the backtest
                    trade['Sell Price'] = np.nan # Indicate no final price
                    trade['Return'] = np.nan # Indicate no final return
            else:
                 # Handle cases where the ticker column doesn't exist in price_df
                 trade['Sell Date'] = last_date.strftime('%Y-%m-%d')
                 trade['Sell Price'] = np.nan
                 trade['Return'] = np.nan


# Convert the list of trade details into a pandas DataFrame
trades_df_output = pd.DataFrame(all_trade_details)

# --- Save all trades to a file ---
trades_df_output.to_csv('all_trades_constrained_portfolio.csv', index=False)
print("All trades saved to 'all_trades_constrained_portfolio.csv'")


# --- Calculate Portfolio Returns based on Simulated Weights ---
# This calculation is based on the portfolio_value list generated in the daily loop
# Ensure portfolio_dates has at least two dates before calculating returns
if len(portfolio_dates) > 1:
    portfolio_returns_series = pd.Series(
        [portfolio_value[i] / portfolio_value[i-1] - 1 for i in range(1, len(portfolio_value))],
        index=portfolio_dates[1:]
    ).fillna(0)
else:
    portfolio_returns_series = pd.Series([], dtype=float) # Empty series if no returns

# Calculate the cumulative equity curve from the portfolio value
equity_constrained = pd.Series(portfolio_value, index=portfolio_dates)

# --- 4) Constrained Portfoliolevel statistics ---
trading_days = 252
# Ensure there are enough data points for meaningful statistics
if len(portfolio_returns_series) > 0:
    years_constrained = len(portfolio_returns_series) / trading_days if trading_days > 0 else 0.001 # Avoid division by zero
    cagr_constrained = equity_constrained.iloc[-1] ** (1/years_constrained) - 1 if len(equity_constrained) > 1 and years_constrained > 0 else 0

    # Sharpe Ratio (Annualized, risk-free rate assumed 0)
    sharpe_constrained = np.sqrt(trading_days) * portfolio_returns_series.mean() / portfolio_returns_series.std() if portfolio_returns_series.std() != 0 else np.nan

    # Sortino Ratio (Annualized, risk-free rate assumed 0)
    downside_returns = portfolio_returns_series[portfolio_returns_series < 0]
    downside_deviation = downside_returns.std() if len(downside_returns) > 0 else 0
    sortino_ratio = np.sqrt(trading_days) * portfolio_returns_series.mean() / downside_deviation if downside_deviation != 0 else np.nan

    # Max Drawdown 1 (Value) and Max Drawdown 2 (Duration)
    # Max Drawdown Value
    roll_max_constrained = equity_constrained.cummax()
    drawdown_series = (equity_constrained - roll_max_constrained) / roll_max_constrained
    max_dd_value = drawdown_series.min() if len(drawdown_series) > 0 else 0

    # Max Drawdown Duration
    if len(drawdown_series) > 0:
        in_drawdown = drawdown_series < 0
        # Handle potential all-positive returns case
        if in_drawdown.any():
            drawdown_id = (in_drawdown & (~in_drawdown.shift(1, fill_value=False))).cumsum()
            drawdown_id = drawdown_id.where(in_drawdown, np.nan)
            # Drop NaNs before grouping
            drawdown_durations = drawdown_id.dropna().groupby(drawdown_id).size()
            max_dd_duration = drawdown_durations.max() if not drawdown_durations.empty else 0
        else:
            max_dd_duration = 0 # No drawdowns
    else:
        max_dd_duration = 0 # Empty drawdown series

else:
    # Default values if no returns data
    years_constrained = 0
    cagr_constrained = 0
    sharpe_constrained = np.nan
    sortino_ratio = np.nan
    max_dd_value = 0
    max_dd_duration = 0


# --- Trade-level statistics (from the saved trades file) ---
closed_trades = trades_df_output[trades_df_output['Status'] == 'Closed']['Return'].dropna() # Ensure no NaNs in returns
win_rate = (closed_trades > 0).mean() * 100 if len(closed_trades) > 0 else 0
avg_profit_per_trade = closed_trades[closed_trades > 0].mean() * 100 if len(closed_trades[closed_trades > 0]) > 0 else 0
avg_loss_per_trade = closed_trades[closed_trades < 0].mean() * 100 if len(closed_trades[closed_trades < 0]) > 0 else 0
expectancy_per_trade = closed_trades.mean() * 100 if len(closed_trades) > 0 else 0 # Avg Return %
# Handle cases where there are no winning or losing trades
total_wins_sum = closed_trades[closed_trades > 0].sum()
total_losses_sum = closed_trades[closed_trades < 0].sum()
if total_losses_sum != 0:
    profit_factor = total_wins_sum / abs(total_losses_sum)
elif total_wins_sum > 0:
    profit_factor = np.inf # Infinite profit factor if no losses and some wins
else:
    profit_factor = 0 # Profit factor is 0 if no wins and no losses or only losses


# --- 5) Summary and Output ---
stats_constrained = pd.Series({
    # --- Trade Stats ---
    "Total Number of Trades": len(trades_df_output),
    "Number of Closed Trades": len(closed_trades),
    "Number of Open Trades": len(trades_df_output[trades_df_output['Status'] == 'Open']),
    "Win Rate %": f"{win_rate:.2f}%",
    "Avg Profit % per Trade": f"{avg_profit_per_trade:.2f}%",
    "Avg Loss % per Trade": f"{avg_loss_per_trade:.2f}%",
    "Expectancy % per Trade": f"{expectancy_per_trade:.2f}%",
    "Profit Factor": f"{profit_factor:.2f}" if not np.isinf(profit_factor) else "Inf",

    # --- Portfolio Stats ---
    "CAGR (Constrained)": f"{cagr_constrained*100:.2f}%",
    "Sharpe Ratio (Constrained)": f"{sharpe_constrained:.2f}" if not pd.isna(sharpe_constrained) else "N/A",
    "Sortino Ratio (Constrained)": f"{sortino_ratio:.2f}" if not pd.isna(sortino_ratio) else "N/A",
    "Max Drawdown % (Value)": f"{max_dd_value*100:.2f}%",
    "Max Drawdown Duration (Days)": f"{max_dd_duration}"
})

print("--- Constrained Portfolio Summary Statistics (with Margin) ---")
print(stats_constrained)


# --- Additional Outputs: Open Positions, Today's Orders, Latest Signals ---
print("\n--- Open Positions ---")
open_positions_df = trades_df_output[trades_df_output['Status'] == 'Open'].copy()
if not open_positions_df.empty:
    # Select and display relevant columns
    display_cols = ['Ticker', 'Buy Date', 'Buy Price', 'Sell Date', 'Sell Price', 'Return']
    # Format the 'Return' column for open trades
    open_positions_df['Return'] = open_positions_df['Return'].apply(lambda x: f"{x*100:.2f}%" if pd.notna(x) else "N/A")
    # Format Buy Price and Sell Price
    open_positions_df['Buy Price'] = open_positions_df['Buy Price'].apply(lambda x: f"{x:.4f}" if pd.notna(x) and x < 1 else (f"{x:.2f}" if pd.notna(x) else "N/A"))
    open_positions_df['Sell Price'] = open_positions_df['Sell Price'].apply(lambda x: f"{x:.4f}" if pd.notna(x) and x < 1 else (f"{x:.2f}" if pd.notna(x) else "N/A"))

    print(open_positions_df[display_cols].to_string(index=False))
else:
    print("No open positions.")


print("\n--- Trades Initiated Today ---")
# Get today's date as a Timestamp
today_timestamp = pd.Timestamp('today').normalize()
# Filter trades that were opened today
todays_trades_df = trades_df_output[
    (trades_df_output['Status'] == 'Open') &
    (pd.to_datetime(trades_df_output['Buy Date']) == today_timestamp) # Convert 'Buy Date' to datetime for comparison
].copy()


if not todays_trades_df.empty:
    display_cols = ['Ticker', 'Buy Date', 'Buy Price']
    # Format Buy Price
    todays_trades_df['Buy Price'] = todays_trades_df['Buy Price'].apply(lambda x: f"{x:.4f}" if pd.notna(x) and x < 1 else (f"{x:.2f}" if pd.notna(x) else "N/A"))
    print(todays_trades_df[display_cols].to_string(index=False))
else:
    print("No new trades initiated today.")


print("\n--- Latest Signals & Indicators ---")
latest_signals_data = []
for t in tickers:
    # Ensure the ticker exists in the signals dictionary and the dataframe is not empty
    if t in signals and not signals[t].empty:
        df = signals[t]
        latest_row = df.iloc[-1]
        # Add checks for column existence before accessing
        # Use .get() with a default value (like np.nan) to avoid KeyErrors
        latest_signals_data.append({
            'Ticker': t,
            'Date': latest_row.name.strftime('%Y-%m-%d'),
            'Close': latest_row.get('Close', np.nan),
            '20-Day High': latest_row.get('HH250', np.nan),
            '5 EMA': latest_row.get('EMA5', np.nan),
            'Signal': latest_row.get('Signal', np.nan)
        })

if latest_signals_data:
    latest_signals_df = pd.DataFrame(latest_signals_data)
    latest_signals_df = latest_signals_df.sort_values(by='Ticker').reset_index(drop=True)
    # Optional: Format numbers for better readability, handling potential NaNs
    latest_signals_df['Close'] = latest_signals_df['Close'].apply(lambda x: f"{x:.4f}" if pd.notna(x) and x < 1 else (f"{x:.2f}" if pd.notna(x) else "N/A"))
    latest_signals_df['20-Day High'] = latest_signals_df['20-Day High'].apply(lambda x: f"{x:.4f}" if pd.notna(x) and x < 1 else (f"{x:.2f}" if pd.notna(x) else "N/A"))
    latest_signals_df['5 EMA'] = latest_signals_df['5 EMA'].apply(lambda x: f"{x:.4f}" if pd.notna(x) and x < 1 else (f"{x:.2f}" if pd.notna(x) else "N/A"))
    latest_signals_df['Signal'] = latest_signals_df['Signal'].apply(lambda x: int(x) if pd.notna(x) else "N/A")

    print(latest_signals_df.to_string(index=False))
else:
    print("No latest signals data available.")


# --- Data Visualization ---
# Plot the equity curve of the constrained portfolio
plt.figure(figsize=(12, 6))
# Ensure equity_constrained is not empty before plotting
if not equity_constrained.empty:
    plt.plot(equity_constrained.index, equity_constrained.values, label='Constrained Portfolio Value')
    plt.xlabel('Date')
    plt.ylabel('Portfolio Value')
    plt.title('Trading Strategy Equity Curve (Constrained with Margin)')
    plt.grid(True)
    plt.legend()
    plt.show()
else:
    print("Not enough data to plot constrained portfolio equity curve.")

# Plot histogram for returns of all trades (closed)
plt.figure(figsize=(12, 6))
if len(closed_trades) > 0:
    plt.hist(closed_trades, bins=50, edgecolor='black')
    plt.xlabel('Return')
    plt.ylabel('Frequency')
    plt.title('Distribution of Closed Trade Returns')
    plt.grid(True)
    plt.show()
else:
    print("Not enough closed trades to plot histogram.")
```

```python
import yfinance as yf
import pandas as pd
import numpy as np
from scipy.stats import skew, kurtosis
import matplotlib.pyplot as plt
import seaborn as sns

# --- 1) Define universe and download data ---
# Define the list of cryptocurrency tickers to include in the analysis
tickers = [
    "BTC-USD","ETH-USD","XRP-USD","USDT-USD","BNB-USD",
    "SOL-USD","USDC-USD","DOGE-USD","TRX-USD","ADA-USD",
    "HYPE-USD","XLM-USD","SUI-USD","LINK-USD","HBAR-USD",
    "BCH-USD","AVAX-USD","SHIB-USD","LTC-USD","LEO-USD",
    "TON-USD","DOT-USD","UNI-USD","XMR-USD","BGB-USD"
]

# Define the date range for data download
# Download data from the start of 2018 up to today for better crypto data availability
endDate = pd.to_datetime('today').strftime('%Y-%m-%d')
# Set group_by to False to avoid a future warning and handle data structure manually
data = yf.download(tickers, start="2010-01-01", end=endDate, group_by=False)

# --- 2) Build signals per coin ---
signals = {}
for t in tickers:
    # Select data for the current ticker and drop rows with any missing values
    # Access data using the ticker as the first level of columns
    df = data[t].copy()

    # Ensure required columns exist and drop rows with NaNs in them
    required_cols = ['Open', 'High', 'Low', 'Close', 'Volume']
    if not all(col in df.columns for col in required_cols):
        print(f"Skipping {t}: Missing one or more required columns ({required_cols}).")
        continue
    df.dropna(subset=required_cols, inplace=True)

    if df.empty:
        print(f"Skipping {t}: No valid data after dropping NaNs.")
        continue

    # Calculate technical indicators and filters
    # Breakout signal: 20-period rolling maximum of the 'Close', shifted by one day
    # Changed from 'High' to 'Close' to match AFL's HHV(C, 250) which was simplified to 20.
    df['HH20'] = df['Close'].rolling(20).max().shift(1)

    # Exponential Moving Averages (EMAs)
    # 5-period EMA of the 'Close' price
    df['EMA5']  = df['Close'].ewm(span=5, adjust=False).mean()
    # Previous day's 5-period EMA
    df['EMA5_p'] = df['EMA5'].shift(1)

    # Bitcoin risk-on filter: True if Bitcoin's close is above its 20-period EMA
    # Get Bitcoin's closing price from the main data DataFrame
    if 'BTC-USD' in data.columns:
        # Ensure BTC-USD data is available and process it
        btc_data = data['BTC-USD'].copy()
        btc_data.dropna(subset=['Close'], inplace=True)
        if not btc_data.empty:
            btc_close_series = btc_data['Close']
            # Calculate Bitcoin's 20-period EMA
            btc_ema20_series = btc_close_series.ewm(span=20, adjust=False).mean()
            # Align Bitcoin's RiskOn status with the current coin's index
            # Use.reindex() and.fillna() to ensure alignment and handle missing dates
            aligned_btc_close = btc_close_series.reindex(df.index, method='ffill')
            aligned_btc_ema20 = btc_ema20_series.reindex(df.index, method='ffill')
            df['RiskOn'] = (aligned_btc_close > aligned_btc_ema20).fillna(False)
        else:
            print(f"Warning: BTC-USD data is empty after dropping NaNs. RiskOn filter disabled for {t}.")
            df['RiskOn'] = True # Default to True if Bitcoin data is unavailable
    else:
        print(f"Warning: BTC-USD data column not found. RiskOn filter disabled for {t}.")
        df['RiskOn'] = True # Default to True if Bitcoin data is unavailable


    # Calculate PositionScore (MA(C*V,21) in AFL)
    # This is used for ranking potential entries
    df['PositionScore'] = (df['Close'] * df['Volume']).rolling(21).mean()
    # Resolve FutureWarning: A value is trying to be set on a copy... using inplace method.
    df['PositionScore'] = df['PositionScore'].fillna(0) # Fill NaNs at beginning of series

    # Generate trading signal: 1=entry, -1=exit, 0=flat
    df['Signal'] = 0
    # Entry signal conditions:
    # 1. Current Close is above the previous day's 20-period Highest High (Breakout)
    # 2. Current Close is above 0.50 (Liquidity filter - changed from 0.20 to 0.50 as per original AFL)
    # 3. Current Volume is above 100,000 (Liquidity filter)
    # 4. Current 5-period EMA is above the previous day's 5-period EMA (Upward EMA trend)
    # 5. Bitcoin RiskOn filter is True
    df.loc[
        (df['Close'] > df['HH20']) &
        (df['Close'] > 0.50) & # Changed from 0.20 to 0.50
        (df['Volume'] > 100_000) &
        (df['EMA5'] > df['EMA5_p']) &
        df['RiskOn'],
        'Signal'
    ] = 1

    # Exit signal condition:
    # Current 5-period EMA is below the previous day's 5-period EMA (Downward EMA trend)
    df.loc[df['EMA5'] < df['EMA5_p'], 'Signal'] = -1

    # Store the DataFrame with signals for the current ticker
    signals[t] = df

# --- 3) Extract trades across all coins with max position limit ---

trade_details = []
open_positions = {}# key: ticker, value: dict with entry info
max_open_trades = 15# <-- HARD CAP
max_wgt_per_trade = 0.10# <-- 10 % of equity (for position sizing)

# Consolidate all signals and relevant data into a single DataFrame for date-by-date processing
# Create a list of DataFrames, each containing 'Close', 'Signal', 'PositionScore' for one ticker
# and rename columns to be unique (e.g., 'BTC-USD_Close', 'BTC-USD_Signal')
all_data_for_backtest = []
for t, df in signals.items():
    if not df.empty:
        # Select only the columns needed for backtesting loop
        temp_df = df[['Close', 'Signal', 'PositionScore']].copy()
        temp_df.columns = [f'{t}_{col}' for col in temp_df.columns]
        all_data_for_backtest.append(temp_df)

if not all_data_for_backtest:
    print("No valid data to run backtest. Exiting.")
    # Create an empty trades_df to avoid errors later
    trades_df = pd.DataFrame(columns=['Ticker', 'Buy Date', 'Buy Price', 'Sell Date', 'Sell Price', 'Return', 'Status'])
    # Skip the rest of the script
    exit()


# Concatenate all dataframes along columns, aligning by date index
portfolio_data = pd.concat(all_data_for_backtest, axis=1)
portfolio_data.sort_index(inplace=True) # Ensure chronological order
# Forward fill Close prices and PositionScores to handle missing data points for some coins on certain days
# Do NOT forward fill Signal, as this should only be active on the exact day it occurs.
close_cols = [col for col in portfolio_data.columns if col.endswith('_Close')]
score_cols = [col for col in portfolio_data.columns if col.endswith('_PositionScore')]
# Resolve FutureWarning: DataFrame.fillna with 'method' is deprecated and will raise in a future version.
portfolio_data[close_cols + score_cols] = portfolio_data[close_cols + score_cols].ffill()
portfolio_data.fillna(0, inplace=True) # Fill any remaining NaNs (e.g., at the very beginning) with 0

# Initialize equity curve and track open positions
margin = 1.5
# Increase equity with margin
equity_curve = [1.0]
equity_dates = [portfolio_data.index[0]]
open_positions = {} # {ticker: {'Buy Date': date, 'Buy Price': price, 'Units': units}}


# Iterate day by day through the consolidated data
for i in range(len(portfolio_data)):
    current_date = portfolio_data.index[i]
    current_day_data = portfolio_data.iloc[i]

    # --- Process Exits for the current day ---
    exited_now = []
    # Iterate over a copy of keys to allow modification during iteration
    for pos_tkr in list(open_positions.keys()):
        exit_signal_col = f'{pos_tkr}_Signal'
        current_close_col = f'{pos_tkr}_Close'

        if exit_signal_col in current_day_data.index and current_close_col in current_day_data.index:
            current_signal = current_day_data[exit_signal_col]
            current_close = current_day_data[current_close_col]

            # Check for exit signal (-1) and ensure the close price is valid
            if current_signal == -1 and not pd.isna(current_close) and pos_tkr in open_positions:
                pos_info = open_positions[pos_tkr]
                ret = (current_close - pos_info['Buy Price']) / pos_info['Buy Price']
                trade_details.append({
                    'Ticker': pos_tkr,
                    'Buy Date': pos_info['Buy Date'].strftime('%Y-%m-%d'),
                    'Buy Price': pos_info['Buy Price'],
                    'Sell Date': current_date.strftime('%Y-%m-%d'),
                    'Sell Price': current_close,
                    'Return': ret,
                    'Status': 'Closed'
                })
                # Update equity based on this closed trade
                # Assuming equity_curve[-1] is the equity *before* processing today's exits/entries
                # A more precise calculation would track position value changes throughout the day.
                # Here, we apply the return of the closed trade to the previous day's equity.
                # This is a simplification for backtesting trade details.
                # The portfolio daily returns calculation (section 5) is a more accurate equity curve.
                # equity_curve.append(equity_curve[-1] * (1 + ret * pos_info['Units'] * (pos_info['Buy Price'] / equity_curve[-1])))
                # equity_dates.append(current_date)
                exited_now.append(pos_tkr)

    # Remove closed positions
    for pos_tkr in exited_now:
        open_positions.pop(pos_tkr, None)


    # --- Process Entries for the current day ---
    entry_candidates = []
    for t in tickers:
        if t == "BTC-USD": # BTC is filter, not traded
            continue

        signal_col = f'{t}_Signal'
        close_col = f'{t}_Close'
        score_col = f'{t}_PositionScore'

        # Ensure columns exist for the current ticker and date
        if signal_col in current_day_data.index and close_col in current_day_data.index and score_col in current_day_data.index:
            current_signal = current_day_data[signal_col]
            current_close = current_day_data[close_col]
            position_score = current_day_data[score_col]

            # Check for entry signal (1), if not already in position, and valid price/score
            if current_signal == 1 and t not in open_positions and not pd.isna(current_close) and current_close > 0 and not pd.isna(position_score):
                entry_candidates.append({
                    'Ticker': t,
                    'Close': current_close,
                    'PositionScore': position_score,
                    'Date': current_date
                })

    # Sort entry candidates by PositionScore (descending)
    # Ensure sorting is done on the PositionScore value
    entry_candidates.sort(key=lambda x: x['PositionScore'], reverse=True)

    # Apply max 10 concurrent positions cap
    num_slots_available = max_open_trades - len(open_positions)
    new_entries_to_take = entry_candidates[:num_slots_available]

    for entry in new_entries_to_take:
        t = entry['Ticker']
        e_idx = entry['Date']
        e_price = entry['Close']

        # Apply max 10% of equity per trade cap
        # Use the latest calculated equity value for sizing
        current_equity = equity_curve[-1]
        # Calculate units based on max_wgt_per_trade
        # This assumes we are buying 'units' of crypto, not shares.
        # If e_price is very small, position_units can be very large.
        if e_price > 0 and current_equity > 0: # Avoid division by zero
            # Calculate the maximum notional value allowed for this trade
            max_notional = max_wgt_per_trade * current_equity
            # Calculate the number of units to buy based on the current price
            position_units = max_notional / e_price
        else:
            position_units = 0 # Cannot open position if price or equity is zero

        if position_units > 0:
            open_positions[t] = {
                'Buy Date': e_idx,
                'Buy Price': e_price,
                'Units': position_units
            }
            # For simplicity in this trade extraction loop, we don't update equity here.
            # The portfolio daily returns calculation (section 5) will handle the overall equity curve.

    # --- Calculate daily portfolio return and update equity curve ---
    # This is a simplified daily return calculation based on open positions.
    # For a real backtester, you would track the value of each position.
    # Here, we'll calculate the weighted average return of open positions.
    daily_returns = []
    total_notional_value = 0
    for pos_tkr, pos_info in open_positions.items():
        current_close_col = f'{pos_tkr}_Close'
        if current_close_col in current_day_data.index and not pd.isna(current_day_data[current_close_col]):
            current_close = current_day_data[current_close_col]
            # Calculate the return for this position since entry
            return_since_entry = (current_close - pos_info['Buy Price']) / pos_info['Buy Price']
            # Calculate the current notional value of the position
            current_notional = current_close * pos_info['Units']
            total_notional_value += current_notional

            # To calculate daily return contribution, we need the previous day's close
            # This requires accessing previous day's data, which is tricky in this loop structure.
            # A simpler approximation for daily return is to assume equal weighting
            # of open positions based on the *initial* allocation logic,
            # but this doesn't reflect actual daily P&L accurately.

            # Let's stick to building the trade list and then calculate portfolio equity
            # separately in section 5, which is designed for this.
            pass # No daily equity update here

    # Append current date to equity_dates for alignment with equity_curve
    if current_date not in equity_dates:
         # This part is not needed if equity_curve is built in section 5
         pass


# --- Handle remaining open positions after the loop ---
# These are trades that were open on the last day of the data
last_date = portfolio_data.index[-1]
for pos_tkr, pos_info in open_positions.items():
    last_price_col = f'{pos_tkr}_Close'
    last_price = portfolio_data.loc[last_date, last_price_col] if last_price_col in portfolio_data.columns and not pd.isna(portfolio_data.loc[last_date, last_price_col]) else pos_info['Buy Price'] # Fallback to buy price if last price is NaN

    ret = (last_price - pos_info['Buy Price']) / pos_info['Buy Price']
    trade_details.append({
        'Ticker': pos_tkr,
        'Buy Date': pos_info['Buy Date'].strftime('%Y-%m-%d'),
        'Buy Price': pos_info['Buy Price'],
        'Sell Date': last_date.strftime('%Y-%m-%d'),
        'Sell Price': last_price,
        'Return': ret,
        'Status': 'Open'
    })

# Convert the list of trade details into a pandas DataFrame
trades_df = pd.DataFrame(trade_details)

# --- 4) Tradelevel statistics ---
# Calculate statistics based only on closed trades for traditional metrics
closed_trades_df = trades_df[trades_df['Status'] == 'Closed'].copy()
trades = closed_trades_df['Return']
wins = trades[trades > 0]
losses = trades[trades < 0]

# Calculate various trade performance metrics
win_pct = len(wins) / len(trades) * 100 if len(trades) > 0 else 0
avg_profit = wins.mean() * 100 if len(wins) > 0 else 0
avg_loss = losses.mean() * 100 if len(losses) > 0 else 0
expectancy = trades.mean() * 100 if len(trades) > 0 else 0
# Profit factor is total winnings divided by total losses (absolute value)
profit_factor = wins.sum() / abs(losses.sum()) if losses.sum() != 0 else np.inf if wins.sum() > 0 else 0

# --- 5) Portfolio daily returns (equalweight among in signal coins) ---
# This section calculates the portfolio equity curve based on the signals.
# It's separate from the trade extraction loop (Section 3) which focuses on individual trades.

# Create a DataFrame of daily signals (1 if in trade, 0 if out based on the signal logic)
# This needs to be reconstructed from the signals dictionary
pos_df = pd.DataFrame({t: signals[t]['Signal'].apply(lambda x: 1 if x == 1 else 0) for t in tickers if t != "BTC-USD" and t in signals and not signals[t].empty})

# Shift positions by one day to align with next-day returns and ensure no lookahead bias
pos_df = pos_df.shift(1).fillna(0) # Fill NaNs (beginning of series) with 0

# Create a DataFrame of daily percentage returns for all relevant tickers
# Use data from Section 1 before any NaNs were dropped specifically for signals
price_df_all = pd.DataFrame({t: data[t]['Close'] for t in tickers if t != "BTC-USD" and t in data.columns and not data[t].empty})
# Resolve FutureWarning: The default fill_method='pad' in DataFrame.pct_change is deprecated
ret_df = price_df_all.pct_change(fill_method=None).shift(-1) # nextday returns

# Align pos_df and ret_df indices and columns
# Use the intersection of indices and columns to ensure alignment
common_indices = pos_df.index.intersection(ret_df.index)
common_columns = pos_df.columns.intersection(ret_df.columns)

pos_df_aligned = pos_df.loc[common_indices, common_columns]
ret_df_aligned = ret_df.loc[common_indices, common_columns]

# Calculate daily portfolio return (sum of returns of held assets divided by number of held assets)
# Avoid division by zero if no positions are held on a given day
port_ret = (ret_df_aligned * pos_df_aligned).sum(axis=1) / pos_df_aligned.sum(axis=1)
# Fill any days with no positions with a 0 return
port_ret = port_ret.fillna(0)

# Calculate the cumulative equity curve
# Start with an initial equity of 1.0
equity = (1 + port_ret).cumprod()
# If equity series is empty, create a base value of 1.0
if equity.empty:
    equity = pd.Series([1.0], index=[portfolio_data.index[0] if not portfolio_data.empty else pd.to_datetime('today')])


# --- 6) Portfoliolevel statistics ---
trading_days = 252 # Approximate number of trading days in a year
# Calculate the number of years covered by the data
years = len(port_ret) / trading_days if len(port_ret) > 0 else 0.001 # Avoid division by zero
# Calculate Compound Annual Growth Rate (CAGR)
# Ensure equity has more than one data point to calculate CAGR
cagr = equity.iloc[-1] ** (1/years) - 1 if len(equity) > 1 and years > 0 else 0
# Calculate Sharpe Ratio (using risk-free rate of 0 for simplicity here, can add later)
# Avoid division by zero if portfolio return standard deviation is 0
sharpe = port_ret.mean() / port_ret.std() * np.sqrt(trading_days) if port_ret.std() != 0 else np.nan
# Calculate Maximum Drawdown
roll_max = equity.cummax()
max_dd = ((equity - roll_max) / roll_max).min() if len(equity) > 0 else 0

# --- 7) Summary and Output ---
# Create a Series to hold the summary statistics
stats = pd.Series({
    "Number of Trades (Closed)": len(closed_trades_df),
    "Number of Trades (Open)": len(trades_df[trades_df['Status'] == 'Open']),
    "Total Number of Trades": len(trades_df),
    "Win % (Closed Trades)": f"{win_pct:.2f}%",
    "Avg Profit % (Closed Trades)": f"{avg_profit:.2f}%",
    "Avg Loss % (Closed Trades)": f"{avg_loss:.2f}%",
    "Expectancy % (Closed Trades)": f"{expectancy:.2f}%",
    "Profit Factor (Closed Trades)": f"{profit_factor:.2f}",
    "Annual Return (CAGR)": f"{cagr*100:.2f}%",
    "Sharpe Ratio": f"{sharpe:.2f}" if not pd.isna(sharpe) else "N/A",
    "Max Drawdown %": f"{max_dd*100:.2f}%"
})
print("--- Summary Statistics ---")
print(stats)

# Print details of open positions
print("\n--- Open Positions ---")
open_trades = trades_df[trades_df['Status'] == 'Open']
if not open_trades.empty:
    display(open_trades[['Ticker', 'Buy Date', 'Buy Price', 'Sell Date', 'Sell Price', 'Return']])
else:
    print("No open positions.")

# Print latest signals from 2024 and 2025
print("\n--- Latest Signals (2024 and 2025) ---")
latest_signals = {}
for t, df in signals.items():
    # Filter for signals in the years 2024 and 2025
    # Ensure df is not empty before attempting to slice
    if not df.empty:
        recent_signals = df.loc[df.index.year.isin([2024, 2025])]
        if not recent_signals.empty:
            # Get the last signal row for the current ticker
            latest_signal_row = recent_signals.iloc[-1]
            # Store the latest signal details
            latest_signals[t] = {
                'Date': latest_signal_row.name.strftime('%Y-%m-%d'),
                'Signal': latest_signal_row['Signal'],
                'Close': latest_signal_row['Close']
            }

# Display the latest signals in a DataFrame if any were found
if latest_signals:
    latest_signals_df = pd.DataFrame.from_dict(latest_signals, orient='index')
    latest_signals_df.index.name = 'Ticker'
    display(latest_signals_df)
else:
    print("No signals found in 2024 and 2025.")

# --- Data Visualization ---
# Plot the equity curve
plt.figure(figsize=(12, 6))
plt.plot(equity)
plt.xlabel('Date')
plt.ylabel('Equity')
plt.title('Trading Strategy Equity Curve')
plt.grid(True)
plt.show()

# Plot histogram for returns of closed trades
plt.figure(figsize=(12, 6))
plt.hist(closed_trades_df['Return'], bins=50, edgecolor='black')
plt.xlabel('Return')
plt.ylabel('Frequency')
plt.title('Distribution of Closed Trade Returns')
plt.grid(True)
plt.show()
```
